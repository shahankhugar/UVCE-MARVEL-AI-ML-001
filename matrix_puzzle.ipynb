```python
import numpy as np
import matplotlib.pyplot as plt

# Load the scrambled matrix
data = np.load('encoded_array.npy')
print("Shape:", data.shape)
print("Total elements:", data.size)
```

```python
# Clue 1 — Reshape into a square
reshaped = data.reshape(100, 100)
print("Reshaped:", reshaped.shape)

# Clue 2 — Data might be sideways
transposed = reshaped.T
print("Transposed:", transposed.shape)

# Clue 3 — End is the beginning (flip it!)
flipped = np.flip(transposed)
print("Flipped:", flipped.shape)
```

```python
%matplotlib inline

plt.figure(figsize=(6, 6))
plt.imshow(flipped, cmap='gray')
plt.title("Decoded Image!")
plt.axis('off')
plt.show()
```

## 🔍 How I Decoded It

The clues guided me step by step:

- **Clue 1** → 10,000 elements → reshaped to 100×100 square
- **Clue 2** → data was sideways → fixed with transpose
- **Clue 3** → end is the beginning → flipped the array

### Operations Used
| Operation | NumPy Code |
|-----------|-----------|
| Reshape | `data.reshape(100, 100)` |
| Transpose | `reshaped.T` |
| Flip | `np.flip(transposed)` |
| Visualize | `plt.imshow(flipped, cmap='gray')` |
